## Back Propagation Neural network with: <br>
- One input layer <br>
- One hidden layer <br>
- One output layer

Importing necessary libraries

```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold
from NeuralNetwork import *
```

Dataset

```python
df = pd.read_csv("housing.csv")
df.isnull().sum()
df.describe()
```

Separating the TRAINING and TESTING columns

```python
X = df.drop(columns = 'MEDV', axis = 1)
y = df['MEDV']

X = X.to_numpy()
y = y.to_numpy()

# NORMALIZE THE DATA
X = (X - np.min(X))/(np.max(X) - np.min(X))
y = (y - np.min(y))/(np.max(y) - np.min(y))
y = y.reshape(-1, 1)

print(X.shape)
print(y.shape)
```

- Take user input
- Call the NeuralNetwork custom module
- Report accuracy/loss for: <br>
        - 5 FOLD CROSS VALIDATION <br> 
        - 10 FOLD CROSS VALIDATION

```python
input_layer_size = X.shape[1]
# No. of neurons in INPUT layers = No. of features
hidden_layer_size = int(input("Enter HIDDEN LAYER SIZE = "))
output_layer_size = int(input("Enter OUTPUT LAYER SIZE = "))
learning_rate = float(input("Enter LEARNING RATE = "))
epochs = 1000

NN = NeuralNetwork(input_layer_size, hidden_layer_size, output_layer_size, learning_rate, epochs)

NN.cross_validation_5(X, y)
NN.cross_validation_10(X, y)
```
